### 6. Modul Buatan Sendiri (Konversi Suhu)

```python
print("Ahmad Fajril Falah")
print("6. Modul Buatan Sendiri\n")

# 1. Membuat file modul konversi.py
script_konversi = """
def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
"""

with open("konversi.py", "w") as f:
    f.write(script_konversi)

# 2. Menggunakan modul tersebut
import konversi

c = 30
f = 86

hasil_f = konversi.celcius_ke_fahrenheit(c)
hasil_c = konversi.fahrenheit_ke_celcius(f)

print(f"{c} Celcius = {hasil_f} Fahrenheit")
print(f"{f} Fahrenheit = {hasil_c} Celcius")
```

### 8. Studi Kasus Mini (Kuis Penjumlahan)

```python
import random

print("Ahmad Fajril Falah")
print("8. Studi Kasus Mini\n")

# 1. Membuat modul hasil.py
script_laporan = """
def tampilkan_skor(nama, skor):
    print("\\n" + "="*30)
    print(f"HASIL NILAI KUIS")
    print("="*30)
    print(f"Nama Peserta : {nama}")
    print(f"Skor Akhir   : {skor}")
    if skor == 100:
        print("Predikat     : wahhh, Sempurna!")
    elif skor >= 70:
        print("Predikat     : okeee, Bagus!")
    else:
        print("Predikat     : yahhh, Belajar Lagi Ya!")
    print("="*30)
"""

with open("hasil.py", "w") as f:
    f.write(script_laporan)

# 2. Program Utama Kuis
import hasil

def mulai_kuis():
    nama = "Fajril"
    skor = 0
    jumlah_soal = 2

    print(f"Halo {nama}, mari mulai kuis penjumlahan!")

    for i in range(jumlah_soal):
        a = random.randint(1, 10)
        b = random.randint(1, 10)
        jawaban_benar = a + b

        # Simulasi soal
        print(f"\nSoal {i+1}: Berapa {a} + {b}?")

        # Jawaban otomatis benar
        tebakan = jawaban_benar
        print(f"Jawabanmu: {tebakan}")

        if tebakan == jawaban_benar:
            print("Benar!")
            skor += 50
        else:
            print(f"Salah! Yang benar adalah {jawaban_benar}")

    # Modul buatan untuk hasilnya
    hasil.tampilkan_skor(nama, skor)

# Jalankan kuis
mulai_kuis()
```
